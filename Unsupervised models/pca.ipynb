```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
```

```python
# load the dataset
wine_data = load_wine()
# split the data into target and features
X = wine_data.data
y = wine_data.target
target_names = wine_data.target_names
```

```python
# Perform PCA to reduce the data to 2 dimensions
pca = PCA(n_components=2)
principle_components = pca.fit_transform(X)
```

```python
# Variance retained
variance_retained = pca.explained_variance_ratio_
```

```python
# Visualize the 2D representation of the dataset using a scatter plot
plt.figure(figsize=(10,7))
scatter = plt.scatter(principle_components[: ,0],principle_components[: ,1],c=y,cmap='Blues_r',edgecolor='k',s=50)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.title('2D PCA Of Wine Dataset')
plt.colorbar(scatter, label='Wine Categories')
plt.grid(True)
plt.show()
variance_retained
```
